# <center>Abstract </center>
Cal Poly Students rate and review their professors on Polyratings.com. The site provides the students anonymity in their individual text reviews, accompanied by the Grade recieved by the student, their year levels, and the title of the course. The site does not publish the ratings given to Professors by individual reviewers; instead Polyratings also provides a Full Statisticial Analysis page for each professor with 5+ reviews/ratings to perform aggregate calculations including mean, median, mode, and standard deviations grouped by the students' grades earned or year levels. 

The following questions are explored through exploratory analysis and machine learning in this report. 

Can we predict the students' overall ratings for a professor given their grades?


### <center>P (Overall rating | Grades, Professor)</center>

Can we predict the students' overall ratings for a professor given the text of their reviews?

# <center> Polyratings Dataset</center>
 
## <center>Professors -- Reviews -- Courses</center>
 

 * **Grades** 
       Grade earned by reviewers 
       (A, B, C, D, F, N/A, Credit, No Credit)'
 * **Year** 
        Year level of the reviewers 
       (Freshman, Sophomore, Junior, Senior, 5th Year Senior, Graduate Student) 
 * **Course**
        Course Dept. and Number
        (ENG 149, DATA 301,... etc.) 
 * **Department**
        Department of the Professor
       (Chemistry and Biochemistry, Statistics, Agribusiness,... etc.)
 * **Review**
       Textual reviews written by students.
       
       
## <center>__Aggregate Data__  </center>
### <center>On a [0.00 to 4.00] scale </center>
 
* **Overall Rating**: 
    Mean rating of a Professor's ability. 
    
    * Overall: Mean, median, mode, standard deviation 
    * by Grade level
    * by Year level
    
    
    
* **Presents Material Clearly**: 
    Mean rating of a Professor's ability to present material clearly
    
    * Overall: Mean, median, mode, standard deviation 
    * by Grade level
    * by Year level
    
    
    
* **Understands Student Difficulties**: 
    Mean rating of a Professor's ability to understand students' difficulties
         
    * Overall: Mean, median, mode, standard deviation 
    * by Grade level
    * by Year level
    
       
* **Reporting Evaluations**
     Number of reviews written for a professor. 
    
    * Overall: Mean, median, mode, standard deviation 
    * by Grade level
    * by Year level
    

## <center> Means of  Organization</center>
* **review_id**

    Many of the statistics are based on the individual sample reviews. These are conveniently accompanied by a 'comment id' in the html of Polyratings.com. 

* **prof_id**

    The professors each have their own review page and full statistical analysis page. Both of the urls for these pages end with a unique number. Let's use this as a professor id.
